```python
df.pivot_table('Respondent Age',columns='gender_of_respondent', index='country')
```

```python
import seaborn as sns
sns.set();
```

```python
with sns.axes_style("white"):
    sns.violinplot(x='country',y='Respondent Age', hue='gender_of_respondent', data=df,
                   palette=['pink','steelblue'])
```

```python
with sns.axes_style("white"):
    sns.jointplot(x='country',y='Respondent Age', data=df)
```

```python
with sns.axes_style("white"):
    sns.displot(data=df, x='Respondent Age')
```

```python
with sns.axes_style("white"):
    sns.histplot(data=df,x='household_size',y='Respondent Age', hue='gender_of_respondent')
```
